## 4.3 Cross-validation

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
batch_size = 64
data_train = tf.data.Dataset.from_tensor_slices((x_train, y_train)).batch(batch_size)
data_test =  tf.data.Dataset.from_tensor_slices((x_test,  y_test )).batch(batch_size)
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10)])
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
model.compile(optimizer="adam", loss=loss_fn, metrics=['accuracy'])
history = model.fit(data_train, epochs=30, validation_data=data_test, verbose = 0)
```

## 4.6 Freezing the coefficients of the lower layer

To a large extent, deep learning is linear regression with random features. To see this, let us consider once again the problem of learning the sine function; cf. problem 2.6.

```python
N = 1000
x_train = np.random.uniform(low=-3, high=3, size=N)
y_train = np.sin(x_train)+np.random.normal(loc=0.0, scale=0.1, size=N)
model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(1)),
  tf.keras.layers.Dense(500, activation='relu'),
  tf.keras.layers.Dense(1)
])
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.MeanSquaredError())
model.fit(x_train, y_train, epochs=100, verbose=0)
plt.plot(np.linspace(-3, 3, 100), model.predict(np.linspace(-3, 3, 100)))
plt.plot(np.linspace(-3, 3, 100), np.sin(np.linspace(-3, 3, 100)), 'r')
plt.show()
```

Now, freeze the coefficients of the bottom layer as explained [here](https://keras.io/api/layers/base_layer/#trainable-property) by setting the trainable property to false. 

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(1)),
  tf.keras.layers.Dense(500, activation='relu'),
  tf.keras.layers.Dense(1)
])
# TODO freeze the coefficients of the bottom layer by setting the trainable property to zero
model.layers[0].trainable = False
model.layers[1].trainable = True

model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.MeanSquaredError())
model.fit(x_train, y_train, epochs=100, verbose=0)
plt.plot(np.linspace(-3, 3, 100), model.predict(np.linspace(-3, 3, 100)))
plt.plot(np.linspace(-3, 3, 100), np.sin(np.linspace(-3, 3, 100)), 'r')
plt.show()
```

You should see a bad training outcome because the bias in the bottom layer is frozen at zero. Thus, all kinks of the piece-wise linear function represented by the network are located at zero. To fix this, randomly initialize the bias as in problem 2.5. You should now see a decent but not great fit.

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(1)),
  tf.keras.layers.Dense(500, activation='relu', bias_initializer=tf.keras.initializers.RandomNormal(mean=0., stddev=1.)),
  tf.keras.layers.Dense(1)
])
# TODO freeze the coefficients of the bottom layer by setting the trainable property to zero
model.layers[0].trainable = False
model.layers[1].trainable = True

model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.MeanSquaredError())
model.fit(x_train, y_train, epochs=100, verbose=0)
plt.plot(np.linspace(-3, 3, 100), model.predict(np.linspace(-3, 3, 100)))
plt.plot(np.linspace(-3, 3, 100), np.sin(np.linspace(-3, 3, 100)), 'r')
plt.show()
```

What have we gained by freezing the bottom-layer coefficients? The number of trainable parameters has been reduced by a factor of 3. Moreover, training has been turned into a linear regression problem, i.e., a linear model with quadratic objective. Such problems can be solved extremely fast by specialized methods (linear least squares via singular value decomposition). The following code should produce the same result but runs much faster.

```python
features = tf.keras.Model(inputs=model.input, outputs=model.get_layer(index=0).output).predict(x_train)
regression = LinearRegression(fit_intercept=False).fit(features, y_train)
model.get_layer(index=1).weights[0] = regression.coef_[:,None]
plt.plot(np.linspace(-3, 3, 100), model.predict(np.linspace(-3, 3, 100)))
plt.plot(np.linspace(-3, 3, 100), np.sin(np.linspace(-3, 3, 100)), 'r')
plt.show()
```

What have we lost by freezing the bottom-layer coefficients? The fit is not as good as before. However, we are not using as many parameters as before, and training is much faster, so we can easily increase the size of the network, i.e., the number of random features. This often leads to suprisingly good results.
